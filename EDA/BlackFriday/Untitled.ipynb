```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#import the train data
df_train = pd.read_csv("train.csv")
df_train.head()
```

```python
df_test = pd.read_csv('test.csv')
df_test.head()
```

```python
# merging the two taining and testing of the data
df = df_train.append(df_test)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#here we take axis=1 beacuse we want to delete the columns 
#if we want to delete the row then axis=0

df.drop(['User_ID'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Gender'] = df['Gender'].map({'F': 0, 'M': 1})
```

```python
df.head()
```

```python
#pd.get_dummies(df['Age'],drop_first=True)
#we can use label encoding
df['Age'] = df['Age'].map({'0-17':1,'18-25':2,'26-35':3,'36-45':4,'46-50':5,'51-55':6,'55+':7})
df.head()
```

```python
df.drop('City_Category',axis=1,inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['Product_Category_2'].unique()
#this is all are the discrete values so replace it with the mode 
#for this never use the mean otherwise it will create the new value instead the replacing it
```

```python
df['Product_Category_2'] = df['Product_Category_2'].fillna(df['Product_Category_2']).mode()[0]
```

```python
df['Product_Category_2'].isnull().sum()
```

```python
df['Product_Category_3'].unique()
```

```python
df['Product_Category_3'] = df['Product_Category_3'].fillna(df['Product_Category_3']).mode()[0]
```

```python
df['Product_Category_2'].isnull().sum()
```

```python
df.head()
```

```python
df.shape
```

```python
df['Stay_In_Current_City_Years'].unique()
```

```python
df['Stay_In_Current_City_Years'] = df['Stay_In_Current_City_Years'].str.replace('+','')
```

```python
df.head()
```

```python
df['Stay_In_Current_City_Years'] = df['Stay_In_Current_City_Years'].astype(int)
```
